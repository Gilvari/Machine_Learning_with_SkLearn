```python
#random forest: multiple random trees
# the important feature about random forest is that. you are able to identifying the important matrics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.datasets import load_iris
from sklearn.datasets import fetch_california_housing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn.tree import export_text
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
data = datasets. load_iris()
df = pd.DataFrame(data.data,columns=data.feature_names) #creating new data frame with iris samples' values
df['target']=data.target #add new column as species which gain the value from the target value of iris dataset
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[data.feature_names], df['target'],  random_state=0)
```

```python
reg=RandomForestClassifier(n_estimators=100,random_state=0)
```

```python
reg.fit(X_train,y_train)
```

```python
print('training accuracy: ',reg.predict(X_test) )
```

```python
reg.score(X_test,y_test)
```

```python
importance_df = pd.DataFrame({'feature': X_train.columns, 'importance': np.round(reg.feature_importances_, 3)})
importance_df = importance_df.sort_values('importance',ascending=False)
```
